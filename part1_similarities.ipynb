```python
#common firend

import jax.numpy as jnp
from jax import grad, jit, vmap
from jax import random

NadjN=np.array(adj)
CF=jnp.dot(NadjN,NadjN).block_until_ready() #matrix multiplication

CF=CF - np.diag(np.diag(CF))#matrix off_diagonal
#print(CF)
CFdf = pd.DataFrame(CF)
with pd.HDFStore('/content/drive/MyDrive/1000/TCFdf.h5',  mode='w') as store:
    store.append('CFdf', CFdf ,format='table')
```

```python
#total_firend

adjc = pd.crosstab(data.src, data.dst)
idx = adjc.columns.union(adjc.index)
up_adj_mtx = adjc.reindex(index = idx, columns=idx, fill_value=0)
up_adj_mtx=up_adj_mtx.to_numpy()

  #print(up_adj_mtx)



i=up_adj_mtx.shape[0]
c=np.zeros(shape=(i,i))
for i in range(0,up_adj_mtx.shape[0]):
  for j in range(0,up_adj_mtx.shape[1]):
    if i==j:
      c[i,j]=(sum(up_adj_mtx[i,:])+sum(up_adj_mtx[:,j]))
  #print(c)
```

```python
#while(False):
i=c.shape[0]
TF=np.zeros(shape=(i,i))
for i in range(0,c.shape[0]):
  for j in range(0,c.shape[1]):
    TF[i,j]=c[i,i]+c[j,j]
    if CF[i,j]!=0:
      TF[i,j]-=CF[i,j]

TF=TF - np.diag(np.diag(TF))#matrix off_diagonal
TFdf = pd.DataFrame(TF)
store = pd.HDFStore('/content/drive/MyDrive/1000/TTFdf.h5')
store['TFdf']=TFdf
store.close()
with pd.HDFStore('/content/drive/MyDrive/1000/TTFdf1.h5',  mode='w') as store:
    store.append('TFdf', TFdf ,format='table')
```

```python
#jaccard_similarity
#while(False):
jaccarddistances = pdist(adj, metric='jaccard')
jaccarddistances = squareform(jaccarddistances)
jaccardsimilarity = 1-jaccarddistances

jaccardsimilarityN=jaccardsimilarity - np.diag(np.diag(jaccardsimilarity))

jaccardsimilaritydfR = pd.DataFrame(jaccardsimilarityN)
with pd.HDFStore('/content/drive/MyDrive/1000/TjaccardsimilaritydfR.h5',  mode='w') as store:
    store.append('jaccardsimilaritydfR', jaccardsimilaritydfR
                  ,format='table')#, data_columns= jaccardsimilaritydfR.columns

  #jaccard_similarity
```

```python
#cosine_similarity
#while(False):
cosine_distances = pdist(adj, metric='cosine')
cosine_distances = squareform(cosine_distances)
cosine_similarity = 1-cosine_distances

cosine_similarityN=cosine_similarity - np.diag(np.diag(cosine_similarity))
cosine_similaritydfR = pd.DataFrame(cosine_similarityN)
store = pd.HDFStore('/content/drive/MyDrive/1000/TcosinesimilaritydfR.h5')
store['cosine_similaritydfR'] = cosine_similaritydfR
store.close()
with pd.HDFStore('/content/drive/MyDrive/1000/TcosinesimilaritydfR1.h5',  mode='w') as store:
    store.append('cosine_similaritydfR', cosine_similaritydfR,format='table')
```

```python
#L1_norm()

#L1_norm()

i=up_adj_mtx.shape[0]
L1_norm1=np.zeros(shape=(i,i))
for i in range(0,L1_norm1.shape[0]):
  for j in range(0,L1_norm1.shape[1]):
    L1_norm1[i,j]=jnp.dot(c[i,i],c[j,j]).block_until_ready()

i=up_adj_mtx.shape[0]
L1_norm=np.zeros(shape=(i,i))
for i in range(0,L1_norm.shape[0]):
  for j in range(0,L1_norm.shape[1]):
    if CF[i,j]==0 or L1_norm1[i,j]==0:
        L1_norm[i,j]=0
    else:
        L1_norm[i,j]=jnp.divide(CF[i,j],L1_norm1[i,j]).block_until_ready()

L1_normN=L1_norm - np.diag(np.diag(L1_norm))#matrix off_diagonal
L1_normdf = pd.DataFrame(L1_normN)
with pd.HDFStore('/content/drive/MyDrive/1000/TL1normdf.h5',  mode='w') as store:
      store.append('L1_normdf', L1_normdf,format='table')
#print(L1_norm)
```

```python
#edge_weight_measure

i=up_adj_mtx.shape[0]
ewm=np.zeros(shape=(i,i))
for i in range(0,ewm.shape[0]):
  for j in range(0,ewm.shape[1]):
     if i==j:
       ewm[i,j]=jnp.divide(1,np.sqrt(1+c[i,i])).block_until_ready()
       #ewm=np.round(ewm)

#print(ewm)
```

```python
i=up_adj_mtx.shape[0]
wsum=np.zeros(shape=(i,i))
for i in range(0,wsum.shape[0]):
  for j in range(0,wsum.shape[1]):
    wsum[i,j]=(ewm[i,i]+ewm[j,j])


wsum=wsum - np.diag(np.diag(wsum))
#print(np.round(wsum,1))

wsumdf = pd.DataFrame(wsum)

with pd.HDFStore('/content/drive/MyDrive/1000/Tsum.h5',  mode='w') as store:
      store.append('wsumdf', wsumdf,format='table')
```

```python
i=up_adj_mtx.shape[0]
wmul=np.zeros(shape=(i,i))
for i in range(0,wmul.shape[0]):
  for j in range(0,wmul.shape[1]):
    wmul[i,j]=jnp.dot(ewm[i,i],ewm[j,j]).block_until_ready()
    

wmul=wmul - np.diag(np.diag(wmul))
#print(np.round(wmul,1))

wmuldf = pd.DataFrame(wmul)

with pd.HDFStore('/content/drive/MyDrive/1000/Tmul.h5',  mode='w') as store:
      store.append('wmuldf', wmuldf,format='table')
```
